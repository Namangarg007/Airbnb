```python
cpy_data.dropna(inplace=True)# dropna removes the rows with empty values if inplace argument is not written 
# then dropna function return a copy of the dataframe and the origianl will stay intact but with dropna
# inplace = True will remove the rows with empty values in the original data
```

```python
# another way to deal with empty value is ti use fillna function it works same as drop na but instead of 
# deleting the rows it fill the rows with empty value with the value provided by default it returns the copy 
# of the original dataframe but by using inplace=True argument it will change the value in original dataframe
```

```python
x = cpy_data['reviews_per_month'].median() # with this we will find the mean of all the values in the specified column
```

```python
cpy_data['reviews_per_month'].fillna(x, inplace=True) # it will fill the empty values of review_per_month
# with the mode value that we have calculated
```

```python
cpy_data.drop('last_review', axis=1, inplace=True) # this will remove the row or column specified by default 
# it will remove the row but by pasing the axis =1 it will remove the column specified
```

```python
cpy_data[['name', 'host_name', 'neighbourhood', 'neighbourhood_group', 'room_type']] = cpy_data[['name', 'host_name', 'neighbourhood', 'neighbourhood_group', 'room_type']].convert_dtypes(str)
```

```python
# to check for duplicate values we use duplicate method

dup = cpy_data.duplicated() #with two string method it will diaplay all the values of data set instead of first 5 and
#last five
```

```python
# since the data set is too big we will check the value the duplicates using for loop
count = 0

for x in dup.index:# this will give the index of the data 
    if dup.loc[x] == True:
        count += 1
print(count)
```

```python
# as we can see there are no duplicate values in the dataframe
# if duplicates were found then we would have used drop_duplicate() function
```
